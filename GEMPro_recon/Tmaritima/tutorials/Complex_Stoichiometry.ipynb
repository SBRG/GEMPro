Then, we analyze data on protein abundances in the context of protein complex stoichiometry: using constraint-based methods, we compute feasible enzyme abundances and identify proteins that are not produced stoichiometrically with their corresponding complexes.

## 3. Construction of enzyme stoichiometry matrix

We will now structure the data into an enzyme stoichiometry matrix. In this, rows are peptides, columns are enzymes and the entries are protein stoichiometry within an enzyme. This will allow for simple queries and also allow computation of enzyme states in section 4.4. For convention, we will store genes with unannotated stoichiometry as having stoichiometry 0 (where as those that do not contain the peptide at all will be annotated as 'NaN').

```python
def parse_pro_stoich(pro_stoich_string):
    pro_stoich_dict={}
    if 'AND' in str(pro_stoich_string):
        pro_strings=str(pro_stoich_string).strip().split(' AND ')
    elif pd.notnull(pro_stoich_string):
        pro_strings=[str(pro_stoich_string)]
    else: 
        pro_strings=['']
    for pro_string in pro_strings:
        if pro_string != '':
            gene,stoich=pro_string.split('(')
            if stoich==')':
                pro_stoich_dict[gene]=0
            else:
                pro_stoich_dict[gene]=int(stoich[:-1])
    return pro_stoich_dict

enzyme_composition_dict=dict([(DF_TM_complexes_update.e_enzyme.loc[ind],\
                               parse_pro_stoich(DF_TM_complexes_update.e_enzyme_composition_string.loc[ind]))\
                              for ind in DF_TM_complexes_update.index])

enzyme_composition_df=pd.DataFrame(enzyme_composition_dict)
```

### 3.2 Exercise

How many enzymes have complete, partial, assumed, or no annotated stoichiometry?

```python
complete_stoichiometry=[]
partial_stoichiometry=[]
no_stoichiometry=[]
for enzyme in enzyme_composition_df:
    genes=enzyme_composition_df[enzyme].dropna()
    
    is_incomplete=False
    has_some_annotation=False
    if min(genes.values)==0:
        is_incomplete=True
    if max(genes.values)!=0:
        has_some_annotation=True
        
    if (has_some_annotation and (not is_incomplete)):
        complete_stoichiometry.append(enzyme)
    elif (has_some_annotation and is_incomplete):
        partial_stoichiometry.append(enzyme)
    else:
        no_stoichiometry.append(enzyme)
        
        
fig, axes = plt.subplots(nrows=1, ncols=1,figsize=(5, 6))

width=0.5

plt.bar(np.arange(3),[len(complete_stoichiometry),len(partial_stoichiometry),len(no_stoichiometry)],width)

#axes.set_title('Completeness of enzyme stoichiometry annotation')
axes.set_ylabel('Number of enzymes')
axes.set_xticks(np.arange(3)+width/2)
axes.set_xticklabels(['complete','partial','none'])
plt.show()

enzyme_composition_complete_df=enzyme_composition_df[complete_stoichiometry]
percent=float(np.true_divide(len(enzyme_composition_df.columns),478))
print "%.1f%% of enzymes have complete stoichiometry"%(100*percent)
```

### 3.3 Exercise 


What fraction are of enzymes are homomers versus heteromers?


```python
enzyme_to_num_genes={}
enzyme_to_num_proteins={}
for enzyme in enzyme_composition_complete_df:
    genes=enzyme_composition_complete_df[enzyme].dropna()
    
    enzyme_to_num_genes[enzyme]=len(genes)
    enzyme_to_num_proteins[enzyme]=int(sum(genes))
    

homomers=[e for e,g in enzyme_to_num_genes.items() if g==1]
heteromers=[e for e,g in enzyme_to_num_genes.items() if g>1]

print '%.1f%% of enzymes are homomers'%(100*float(len(homomers))/(len(homomers)+len(heteromers)))
print '%.1f%% of enzymes are heteromers'%(100*float(len(heteromers))/(len(homomers)+len(heteromers)))
```
