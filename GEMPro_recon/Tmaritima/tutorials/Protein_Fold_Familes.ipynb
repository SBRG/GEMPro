# Understanding evolutionary relationships of fold families in metabolism

This is an example of mapping GEM gene IDs to their respective Protein Fold Family (PFAM) IDs, given the available PDB file(s). In this tutorial, we perform this mapping for T. maritima

The end goal is to establish the coverage of available PFAM annotations for a given M model and where there are missing PFAM entries, run the HMMER source code (http://hmmer.janelia.org) to fill in the gaps.

Available information about the PFAM database can be found here:

- http://pfam.xfam.org/  (UK)
- http://pfam.janelia.org/  (USA)
- Finn, Robert D., et al. "Pfam: the protein families database." Nucleic acids research (2013): gkt1223.

Pfam data are available in a variety of formats, which include flatfiles (derived from the MySQL database) and relational table dumps, both of which can be downloaded from the FTP site (ftp://ftp.sanger.ac.uk/pub/databases/Pfam). The Pfam website (available at http://pfam.sanger.ac.uk/and http://pfam.janelia.org/) provides different ways to access the database content, providing both graphical representations of and interactive access to the data.

In this tutorial, we will be adding PFAM-related information to the Master GEM-PRO dataframe. The following codes are used to track where information is coming from: 

* Columns starting with:
    * m_ is information that comes directly from the GEM
    * u_ is information that comes from UniProt
    * p_ is information that comes from the PDB
    * i_ is information that comes from I-tasser (or Zhang group)
    * f_ is information that comes from PFAM online database
    * h_ is information that comes from HMMer code
    * ssb_ is information manually annotated

```python
from Bio.Seq import Seq
from Bio.SeqRecord import SeqRecord
from Bio.Alphabet import generic_protein
from Bio import SeqIO

def write_rec_to_fasta(seq_string, id_string, desc_string, file_name):

    rec1 = SeqRecord(Seq(seq_string, generic_protein),
                     id=id_string,
                     description=desc_string)
               
    my_records = [rec1]
    #Now we have a SeqRecord object, we’ll write it to a FASTA format file:

    SeqIO.write(my_records, file_name, "fasta")
```

### 3.3 Add PFAM and CLAN information to GEM-PRO master data frame

We now combine all of the above information into a single dataframe column in GEM-PRO for future studies. 
We keep track of where it comes from in a comments column:
    * "from source" accounts for PFAM data coming from the source code 
    * "from official" accounts for PFAMs that are officially annotated

```python
df1 = DF_04_2_RXN_GENE_UNIPROT_PDB_HM_PFAM[pd.notnull(DF_04_2_RXN_GENE_UNIPROT_PDB_HM_PFAM.f_pfam_acc)][['m_gene','m_reaction','f_pfam_acc']].drop_duplicates()
df1['ssb_pfam_source'] = "from official"
pfams_in_TM = [i[0] for i in df1['f_pfam_acc'].str.split('.')]

df2 = DF_04_2_RXN_GENE_UNIPROT_PDB_HM_PFAM[pd.notnull(DF_04_2_RXN_GENE_UNIPROT_PDB_HM_PFAM.h_hmm_acc)]
df2['h_hmm_acc'] = [i[0] for i in df2['h_hmm_acc'].str.split('.')]
df2 = df2[~df2.h_hmm_acc.isin(pfams_in_TM)][['m_gene','m_reaction','h_hmm_acc']].drop_duplicates()
df2.rename(columns={'h_hmm_acc': 'f_pfam_acc'}, inplace=True)
df2['ssb_pfam_source'] = "from source"

df3 = pd.concat([df1,df2])

read_to_struct = []
for i in df3.index:
    react = df3.m_reaction[i]
    sys = m.reactions.get_by_id(str(react)).subsystem
    pfam = df3.f_pfam_acc[i].split('.')[0]
    if pfam in DF_clans.pfam_id.tolist():
        cl = DF_clans[DF_clans.pfam_id == pfam].clan_name.values[0]
        pf_name = DF_clans[DF_clans.pfam_id == pfam].pfam_name.values[0]
        if str(cl) == '\N':
            cl = 'None'
    else:
        cl = 'None'
    read_to_struct.append({'ssb_pfam_source':df3.ssb_pfam_source[i],'m_subsystem':sys,'m_gene':df3.m_gene[i], 'm_reaction':react,'ssb_pfam':pfam,'f_clan':cl,'f_pfam_name':pf_name})

DF_GEM_PRO_PFAM = pd.DataFrame(read_to_struct)
```
