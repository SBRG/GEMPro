###2.2 Match <i>T. maritima</i> PDB structures to PDB Ligands

This section links each pdb structure in our GEMPRO to the PDB ligand data in order to determine whether each structure has a ligand bound

```python
mol_to_struct =[]
for i in cc_to_pdb.index:
    pdbs_upper=[]
    pdbs = set(cc_to_pdb.ix[i].pdbs.split(' ')[:-1])
    for pdb in pdbs:
        pdbs_upper.append(pdb.upper())
    cc = cc_to_pdb.ix[i].cc
    intersect = set(pdbs_upper) & list_pdbs
    if len(intersect)>0:
        pdbs_out = ' '.join(list(intersect))
        #print cc, pdbs_out, len(intersect)
        for pdb in list(intersect):
            mol_to_struct.append({'pdb':pdb,'cc':cc})
mol_to_struct = pd.DataFrame(mol_to_struct)
```

####Find PDB interactions (example: http://www.pdb.org/pdb/images/FFO:1DFO_psv_v_500.png)

```python
def check_for_interaction(pdb, mol): # check to see if the image file exists or not...
    url = 'http://www.pdb.org/pdb/images/'+mol+':'+pdb+'_sv_v_500.png'
    req = urllib2.Request(url)
    f = urllib2.urlopen(req)
    result = f.read()
    if result.startswith('GIF'):
        return '-'
    else:
        return url
check_for_interaction('1DFO','FFO')
```

##3.1 Matching PDB to EC Number

This section usees the RCSB to map PDB data to Enzyme Commission (EC) numbers.  These numbers can then be used to classify the types of metabolic reactions captured by PDB structures.

```python
def get_pdb_ec(pdb):
    import urllib2
    from bs4 import BeautifulSoup
    url = 'http://www.rcsb.org/pdb/rest/describeMol?structureId='+pdb
    req = urllib2.Request(url)

    f = urllib2.urlopen(req)

    result = f.read()

    soup = BeautifulSoup(result)
    out = []
    for e in soup.findAll('enzclass'):
        out.append(e.attrs['ec'])
    return out
get_pdb_ec('1ILD')
```

## 4 Anaylze Structural Differences Between Ligands

This section uses open babel to calculate pairwise similarity of all ligands in the GEMPRO.  This similarity can then be compared for ligands bound to different genes and the reactions they catalyze

```python
def obcompare(mol1,mol2): #use open babel to compare two ligands
    dir = '/home/ebrunk/Dropbox/SBRG/gempro/SBRG_server/iJO1366/assessment_ligand_binding/sdf_files/'
    mol1 = dir + mol1 + '.sdf'
    mol2 = dir + mol2 + '.sdf'
    result = !babel $mol1 $mol2 -ofpt
    for r in result:
        if r.find('Tanimoto')>=0:
            tanimoto = r.split(' = ')[1]
            try:
                tanimoto = float(tanimoto)
            except:
                print error, tanimoto
                tanimoto = 0
            return tanimoto
obcompare('ATP','EDO')
```

####Calculate pairwise similarity between all compounds that bind to a given protein

```python
gene_cc = gene_cc.dropna()
similarity_matrices = {}
for g in set(gene_cc.m_gene.tolist()):
    mols = gene_cc[gene_cc.m_gene==g]['cc'].tolist()
    #print mols
    #print g
    count = 0
    score = 0
    matrix = pd.DataFrame(columns=mols, index=mols)
    #print matrix
    for m1 in range(0,len(mols)):
        for m2 in range(0,len(mols)):
            if m1==m2:
                tanimoto=1
            mol1 = matrix.index[m1]
            mol2 = matrix.columns[m2]
            val1 = matrix.ix[mol1][mol2]
            val2 = matrix.ix[mol2][mol1]
            if str(val1) == 'nan':
                tanimoto = obcompare(mol1, mol2)
                #print mol1, mol2, tanimoto
                matrix.ix[mol1][mol2] = tanimoto
                matrix.ix[mol2][mol1] = tanimoto
    similarity_matrices[g] = matrix
dump(similarity_matrices,open('met_similarity_matrices_TM.pickle','w'))
```

```python
rxn_tanimotos = []
for r in rxn_similarity:
    matrix = rxn_similarity[r]
    max = matrix.replace(1,0).max().max()
    min = matrix.replace(0,1).min().min()
    if min == 1:
        min = 0
    rxn_tanimotos.append({'reaction':r,'min': min, 'max':max})
rxn_tanimotos = pd.DataFrame(rxn_tanimotos)
```
