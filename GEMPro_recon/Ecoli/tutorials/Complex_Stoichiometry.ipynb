```python
import sys, time
try:
    from IPython.display import clear_output
    have_ipython = True
except ImportError:
    have_ipython = False

class ProgressBar:
    def __init__(self, iterations):
        self.iterations = iterations
        self.prog_bar = '[]'
        self.fill_char = '*'
        self.width = 40
        self.__update_amount(0)
        if have_ipython:
            self.animate = self.animate_ipython
        else:
            self.animate = self.animate_noipython

    def animate_ipython(self, iter):
        print '\r', self,
        sys.stdout.flush()
        self.update_iteration(iter + 1)

    def update_iteration(self, elapsed_iter):
        self.__update_amount((elapsed_iter / float(self.iterations)) * 100.0)
        self.prog_bar += '  %d of %s complete' % (elapsed_iter, self.iterations)

    def __update_amount(self, new_amount):
        percent_done = int(round((new_amount / 100.0) * 100.0))
        all_full = self.width - 2
        num_hashes = int(round((percent_done / 100.0) * all_full))
        self.prog_bar = '[' + self.fill_char * num_hashes + ' ' * (all_full - num_hashes) + ']'
        pct_place = (len(self.prog_bar) // 2) - len(str(percent_done))
        pct_string = '%d%%' % percent_done
        self.prog_bar = self.prog_bar[0:pct_place] + \
            (pct_string + self.prog_bar[pct_place + len(pct_string):])

    def __str__(self):
        return str(self.prog_bar)
```

### 2.1 Load protein stoichiometry and reaction-enzyme mapping files


This data is obtained from the protein complex reconstruction performed in:

* O'Brien E.J., Lerman J.A., Chang R.L., Hyduke D.R., Palsson B.O. Genome-scale models of metabolism and gene expression extend and refine growth phenotype prediction. Mol Syst Biol., 9:693. (2013).

Which is based primarily on annotations present in EcoCyc database:

* Keseler, I.M., Mackie, A., Peralta-Gil, M., Santos-Zavaleta, A., Gama-Castro, S., Bonavides-Martinez, C., Fulcher, C., Huerta, A.M., Kothari, A., Krummenacker, M., Latendresse, M., Muniz-Rascado, L., Ong, Q., Paley, S., Schroder, I., Shearer, A., Subhraveti, P., Travers, M., Weerasinghe, D., Weiss, V., Collado-Vides, J., Gunsalus, R.P., Paulsen, I., and Karp, P.D., 
EcoCyc: fusing model organism databases with systems biology
Nucleic Acids Research 41:D605-612 2013.

These are obtained by first mapping gene sets in GPRs that are connected by 'and' relationships to protein complexes in the EcoCyc database. In cases where the gene composition and the reaction catalyzed in EcoCyc annotation matches exactly with the metabolic model, the protein stoichiometry annocated in EcoCyc is used. In these cases, enzyme IDs are identical to those in EcoCyc. However, in cases where the set of proteins necessary for catalysis do not match the complexes annotated in EcoCyc (e.g. if there is a regulatory protein subunit), a new ID is created (that is similar but not identical to that used in EcoCyc).

Only metabolic protein complexes are included in the PROTEIN_COMPLEX_DATA file. Protein complexes relevant for catalysis of protein expression reactions (e.g. transcription, translation, RNA modification, protein folding) can be found in the supplemental material of O'Brien et al.

In the original reconstruction in O'Brien et al. enzyme complexes are named and annotated with metal ions and prosthetic groups necessary for catlysis. Here, only the names of the protein complexes (without additional modifications) are included.

## 3. Construction of enzyme stoichiometry matrix

We will now structure the data into an enzyme stoichiometry matrix. In this, rows are peptides, columns are enzymes and the entries are protein stoichiometry within an enzyme. This will allow for simple queries and also allow computation of enzyme states in section 4.4. For convention, we will store genes with unannotated stoichiometry as having stoichiometry 0 (where as those that do not contain the peptide at all will be annotated as 'NaN').

```python

def parse_pro_stoich(pro_stoich_string):
    pro_stoich_dict={}
    pro_strings=pro_stoich_string.strip().split(' AND ')
    for pro_string in pro_strings:
        gene,stoich=pro_string.split('(')
        if stoich==')':
            pro_stoich_dict[gene]=0
        else:
            pro_stoich_dict[gene]=int(stoich[:-1])
    return pro_stoich_dict

enzyme_composition_dict=dict([(complexes_data.complex_name.loc[ind],\
                               parse_pro_stoich(complexes_data.protein_stoichiometry.loc[ind]))\
                              for ind in complexes_data.index])

enzyme_composition_df=pd.DataFrame(enzyme_composition_dict)
```

## 4 Compute feasible enzyme states


With the enzyme stoichiometry matrix and quantitative protein abundance data, we can compute feasible enzyme states.

The basic concept is similar to flux balance analysis commonly used to analyze metabolic networks. Using the enzyme stoichiometry matrix, the feasible abundances of enzymes and the partitioning of proteins among enzymes can be computed.

We will perform several analysis using constraint-based analysis and linear programming with COBRApy:

*  Ebrahim A., Lerman J.A., Palsson B.O., Hyduke D.R. COBRApy: COnstraints-Based Reconstruction and Analysis for Python. BMC Syst Biol 7:74 (2013).


#####Only consider:
    *1) genes with complete stoichiometry across all enzymes, and
    *2) enzymes with complete stoichiometry across all genes

```python

genes_with_partial_stoichiometry=enzyme_composition_df[partial_stoichiometry+no_stoichiometry].sum(axis=1).dropna().index

genes_with_full_stoichiometry=[i for i in enzyme_composition_df.index if i not in genes_with_partial_stoichiometry]

enzyme_composition_subset=enzyme_composition_df[complete_stoichiometry].loc[genes_with_full_stoichiometry]

```

Create Model in COBRApy with enzyme composition matrix

```python
enzyme_model=Model()

#create 'Metabolite' object for each peptide; these will be the rows in the matrix
mets=[]
for pep in enzyme_composition_subset.index:
    mets.append(Metabolite(id=pep))
enzyme_model.add_metabolites(mets)

#create 'Reaction' object for each enzyme; these will be the columns in the matrix
rxns=[]
for enz in enzyme_composition_subset.columns:
    rxn=Reaction(enz)
    rxn.upper_bound=10**10
    rxn.add_metabolites(dict([(enzyme_model.metabolites.get_by_id(k),float(v)) \
                             for k,v in enzyme_composition_subset[enz].dropna().to_dict().items()]))
    rxns.append(rxn)
enzyme_model.add_reactions(rxns)
```

Now we will load in quantitative protein expression data to set protein expression levels.

This is the protein abundance in MOPS minimal media inferred from ribosome profiling data in:
* Li GW, Burkhardt D, Gross C, Weissman JS. (2014) Quantifying absolute protein synthesis rates reveals principles underlying allocation of cellular resources. Cell, 157: 624-35.


```python

protein_abundance=pd.read_csv('protein_abundance.csv',index_col=0)

for pep in enzyme_model.metabolites:
    if pep.id in protein_abundance.index:
        pep._bound=protein_abundance.loc[pep.id].protein_abundance
    else:
        #print pep.id #now, some enzymes might have no peptides
        pep.remove_from_model()
        
#add free peptide components
free_peps=[]
for pep in enzyme_model.metabolites:
    rxn=Reaction('free_'+pep.id)
    rxn.upper_bound=10**10
    rxn.add_metabolites({pep:1.})
    free_peps.append(rxn)
enzyme_model.add_reactions(free_peps)
```

```python
#remove enzymes with missing peptides

rxn_to_remove=[]
for rxn in enzyme_model.reactions:
    remove_rxn=True
    for met in rxn._metabolites.keys():
        try:
            enzyme_model.metabolites.get_by_id(met.id)
            remove_rxn=False
        except:
            print rxn.id
    if remove_rxn:
        rxn_to_remove.append(rxn)
enzyme_model.remove_reactions(rxn_to_remove)
```

### 4.1 Excercise

What proteins are not produced stoichiometrically?

To answer this question, we will find the minimum abundance of all free peptides. (This may take a couple of minutes to run.)

```python
minimum_free_pep_abundance={}
p = ProgressBar(len(free_peps))
for ind,pep in enumerate(free_peps):
    p.animate(ind+1)
    
    enzyme_model.change_objective(pep.id)
    enzyme_model.optimize('minimize')
    minimum_free_pep_abundance[pep.id]=enzyme_model.solution.f
```

### 4.2 Excercise


What are the possible enzyme abundances?

In particular what are the maximal and minimal abundances for each enzyme?

First, we compute maximal abundances.


```python
maximum_enzyme_abundance={}
p = ProgressBar(len([i for i in enzyme_model.reactions if i not in free_peps]))
for ind,enz in enumerate([i for i in enzyme_model.reactions if i not in free_peps]):
    p.animate(ind+1)
    
    enzyme_model.change_objective(enz.id)
    enzyme_model.optimize('maximize')
    maximum_enzyme_abundance[enz.id]=enzyme_model.solution.f
```


What are minimal enzyme abundances assuming free peptide abundance is minimized?


```python

for rxn in enzyme_model.reactions:
    rxn.objective_coefficient=0
for pep in free_peps:
    pep.objective_coefficient=1
enzyme_model.optimize('minimize')
min_free_pep=enzyme_model.solution.f
min_free_pep_constraint=Metabolite(id='min_free_pep_constraint')
min_free_pep_constraint._bound=min_free_pep
enzyme_model.add_metabolites([min_free_pep_constraint])
for pep in free_peps:
    pep.add_metabolites({min_free_pep_constraint:1})

minimum_enzyme_abundance={}
p = ProgressBar(len([i for i in enzyme_model.reactions if i not in free_peps]))
for ind,enz in enumerate([i for i in enzyme_model.reactions if i not in free_peps]):
    p.animate(ind+1)
    
    enzyme_model.change_objective(enz.id)
    enzyme_model.optimize('minimize')
    minimum_enzyme_abundance[enz.id]=enzyme_model.solution.f
```
